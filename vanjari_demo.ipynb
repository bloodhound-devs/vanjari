## Get the Bacteriophage MS2 (Emesvirus zinderi) genome

```python
from torchapp.download import cached_download

url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=nuccore&id=NC_001417.2&rettype=fasta&retmode=text"
filename = "MS2.fasta"
cached_download(url, filename)
```

## Run VanjariFast on the MS2 genome

The first time this runs, it will download the model weights. This may take some time.

```python
from vanjari.apps import VanjariFast

vanjari_fast = VanjariFast()
vanjari_fast(input=filename, output_csv="vanjari-fast-MS2.csv", image_dir="vanjari-fast-MS2-images")
```

## Run Vanjari on the MS2 genome

The first time this runs, it will download the model weights. This may take some time.

This may require more GPU memory than is typically available on a free Google Colab runtime. It is best to run this on an A100 or better.

```python
from vanjari import Vanjari

vanjari = Vanjari()
vanjari(input=filename, output_csv="vanjari-MS2.csv", image_dir="vanjari-MS2-images")
```
