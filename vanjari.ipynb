#Vanjari

Virus Assessment using Neural networks for Just-in-time Analysis and Rapid Identification

[Vanjari](https://github.com/bloodhound-devs/vanjari) is a tool for the classification of viruses using neural networks. It uses the ICTV taxonomy to hierarchically classify viruses.

* **Before you start, please make sure you change the runtime to T4 GPU (or any other kind of GPU if you have $$$), otherwise Phold won't be installed properly**
* To do this, go to the top toolbar, then to Runtime -> Change runtime type -> Hardware accelerator

* To run the cells, press the play button on the left side
* Cell 1 installs vanjari.
* Cell 2 runs vanjari.


```python
#@title 1. Install vanjari

#@markdown This cell installs vanjari. It will take a few minutes. Please be patient

%%bash

set -e

PYTHON_VERSION="3.11"

echo "python version ${PYTHON_VERSION}"

if [ ! -f CONDA_READY ]; then
  echo "installing python"
  wget -qnc https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh
  bash Miniconda3-latest-Linux-x86_64.sh -bfp /usr/local 2>&1 1>/dev/null
  rm Miniconda3-latest-Linux-x86_64.sh
  conda config --set auto_update_conda false
  touch CONDA_READY
fi

if [ ! -f VANJARI_READY ]; then
  echo "installing vanjari"
  conda install -y -c conda-forge -c bioconda pip python=${PYTHON_VERSION}
  pip install git+https://github.com/bloodhound-devs/vanjari.git
  touch VANJARI_READY
fi



```

```python
#@title 3. Run Vanjari

#@markdown First, upload your viruses(s) as a nucleotide input FASTA file

#@markdown Click on the folder icon to the left and use the file upload button.

#@markdown Once it is uploaded, write the file name in the INPUT_FILE field on the right.

#@markdown The file to download is virus-predictions.csv

#@markdown If you do not see the output file,
#@markdown refresh the window by either clicking the folder with the refresh icon below "Files"
#@markdown or double click and select "Refresh".


%%time
import os
import sys
import subprocess
import zipfile
INPUT_FILE = '' #@param {type:"string"}

if os.path.exists(INPUT_FILE):
    print(f"Input file {INPUT_FILE} exists")
else:
    print(f"Error: File {INPUT_FILE} does not exist")
    print(f"Please check the spelling and that you have uploaded it correctly")
    sys.exit(1)


# Construct the command
command = f"vanjari --input {INPUT_FILE} --output-csv  virus-predictions.csv"


# Execute the command
try:
    print("Running vanjari")
    subprocess.run(command, shell=True, check=True)
    print("vanjari completed successfully.")
    print(f"Your output is virus-predictions.csv.")

except subprocess.CalledProcessError as e:
    print(f"Error occurred: {e}")





```
